```python
import pandas as pd
import nltk
import re
from nltk.tokenize import word_tokenize
from collections import Counter
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess_title(title):
    title = title.lower()
    title = re.sub(r'[^a-zA-Z\s]', '', title)
    tokens = word_tokenize(title)
    tokens = [token for token in tokens if token not in stop_words]
    return tokens

all_tokens = []

for title in data['Title']:
    tokens = preprocess_title(title)
    all_tokens.extend(tokens)

token_counts = Counter(all_tokens)
top_15_tokens = token_counts.most_common(15)

print("Top 15 từ xuất hiện nhiều nhất:")
for token, count in top_15_tokens:
    print(f"{token}: {count}")
```

```python
token_label_pairs = []

for title, label in zip(data['Title'], data['class']):
    tokens = preprocess_title(title)
    for token in tokens:
        token_label_pairs.append((token, label))

token_label_df = pd.DataFrame(token_label_pairs, columns=['token', 'class'])
token_class_counts = token_label_df.groupby(['token', 'class']).size().reset_index(name='count')
```

```python
pivot_table = token_class_counts.pivot_table(index='token', columns='class', values='count', fill_value=0)
top_15_tokens = token_class_counts.groupby('token')['count'].sum().nlargest(15).index
pivot_table_top_15 = pivot_table.loc[top_15_tokens]

print(pivot_table_top_15)
```

```python
total_counts = pivot_table_top_15.sum(axis=1)
percentages = (pivot_table_top_15.T / total_counts * 100).T
print(percentages)
```

```python
plt.figure(figsize=(10, 6))  
bottom = None
for label in percentages.columns:
    if bottom is None:
        bottom = percentages[label]
        plt.bar(percentages.index, percentages[label], width = 0.6, label=label.capitalize())
    else:
        plt.bar(percentages.index, percentages[label], width = 0.6, bottom=bottom, label=label.capitalize())
        bottom += percentages[label]

plt.xlabel('Tokens')
plt.ylabel('% of Videos')
plt.title('Top 15 Tokens in Titles by Class')
plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x) + 1) / len(x)
    return x, y

classes = data['class'].unique()

colors = {
    'disturbing': 'blue',
    'irrelevant': 'orange',
    'restricted': 'green',
    'suitable': 'red'
}

line_styles = {
    'disturbing': ':',
    'irrelevant': '-.',
    'restricted': '--',
    'suitable': '-'
}

plt.figure(figsize=(10, 6))

for cls in classes:
    class_data = data[data['class'] == cls]['viewCount']
    x, y = ecdf(class_data)
    plt.plot(x, y, linestyle=line_styles[cls], color=colors[cls], label=cls)

plt.xscale('log')
plt.xlabel('# of views')
plt.ylabel('CDF')
plt.title('CDF of Views per Class')
plt.legend()
plt.grid(True)
plt.show()
```
